# Extracting Annual Revenue Data from 10 of the Top U.S. Stocks by Dollar Volume (Web Scrapping)

```python
# IMPORT LIBRARIES

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Extract the table with the Tesla Annual Revenue Data into the tesla_html variable

tesla_html=requests.get('https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue').text
```

```python
soup = BeautifulSoup(tesla_html,'html5lib')
```

```python
tables = soup.find_all('table')
tesla_revenue = pd.DataFrame(columns=["Date", "Revenue(M)"])
for i,table in enumerate(tables):
    if("Tesla Annual Revenue"in str(table)):
            table_index=i
            for rows in tables[table_index].tbody.find_all("tr"):
                    cols=rows.find_all("td")
                    date=cols[0].text
                    revenue=cols[1].text.replace("$","").replace(",","")
                    tesla_revenue=tesla_revenue.append({"Date":date, "Revenue(M)":revenue}, ignore_index=True)
```

```python
# Removing any rows that contain empty, NAN values

tesla_revenue=tesla_revenue[tesla_revenue["Revenue(M)"] != ""]
```

```python
tesla_revenue.head()
```

```python
# Now to merge the individual company dataframes created above into a single dataframe

from functools import reduce
annual_revenue_dataframes = [tesla_revenue, apple_revenue, amazon_revenue, microsoft_revenue, facebook_revenue, nvidia_revenue, amd_revenue, moderna_revenue,  googl_revenue, affirm_revenue]
master_annual_revenues = reduce(lambda left, right: pd.merge(left,right,on=['Date'], how='outer'), annual_revenue_dataframes)
```

```python
# Now to check the new dataframe

master_annual_revenues
```

```python
# Create new column names for the newly created master_revenues dataframe with the company names

new_column_list=('Date', 'Tesla', 'Apple', 'Amazon', 'Microsoft', 'Facebook', 'NVIDIA', 'AMD', 'Moderna', 'Alphabet', 'Affirm Holdings')
master_annual_revenues.columns = new_column_list
```

```python
# Because these columns are to represent revenue values, they will be converted to float64 and 
# the date column will be converted to datetime

master_annual_revenues=master_annual_revenues.astype({"Tesla":"float64", "Apple":"float64", "Amazon":"float64", "Microsoft":"float64", "Facebook":"float64", "NVIDIA":"float64", "AMD":"float64", "Moderna":"float64", "Alphabet":"float64", "Affirm Holdings":"float64"})
```
